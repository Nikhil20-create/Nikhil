```python
from sklearn.tree import DecisionTreeClassifier
x=[
[60,700],
[50,600],
[40,500],
[30,400],
[20,300],
[10,200]
]
y=[
[1],
[1],
[1],
[1],
[1],
[0]
]
clf=DecisionTreeClassifier()
clf.fit(x,y)
new_data=[[45,100]]
prediction=clf.predict([[40,500]])
if prediction[0]==1:
    print("loan approved") # indented 4 spaces
else:
    print("loan not approved") # indented 4 spaces
```

```python
from sklearn.neighbors import KNeighborsClassifier # Changed kneighborsclassifier to KNeighborsClassifier
import numpy as np
x=np.array([
[7,3],
[6,2],
[4,2],
[2,5],
[5,4],
])
y=['45000','30000','25000','50000','40000']
knn=KNeighborsClassifier(n_neighbors=3) # Changed kneighborsclassifier to KNeighborsClassifier
knn.fit(x,y)
new_number=np.array([[4,2]])
prediction=knn.predict(new_number)
print(f"the overall is:{prediction[0]}")
```

```python
from sklearn import svm
import numpy as np
x=np.array([
[150,9],
[180,7],
[200,8],
[220,8],
])
y=np.array([0,0,1,1])
clf=svm.SVC(kernel='linear')
clf.fit(x,y)
new_fruit=np.array([[160,6]])
prediction=clf.predict(new_fruit)
if prediction[0]==0:
    print("apple") # indented 4 spaces
else:
    print("orange") # indented 4 spaces
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=300,centers=2,cluster_std=0.60,random_state=0)
plt.scatter(x[:,0],x[:,1],s=50)
plt.title("customer data")
plt.xlabel("annual income (k$)")
plt.ylabel("spending score(1-100)")
plt.show()
```

```python
kmeans=KMeans(n_clusters=2)
kmeans.fit(x)
centers=kmeans.cluster_centers_
labels=kmeans.predict(x) # Changed kmeans_predict to kmeans.predict
y_pred=kmeans.predict(x)
plt.scatter(x[:,0],x[:,1],c=y_pred,s=50,cmap='viridis')
plt.scatter(centers[:, 0],centers[:, 1],c='black',s=200,alpha=0.5,marker='x')
plt.title("customer data")
plt.xlabel("annual income (k$)")
plt.ylabel("spending score(1-100)")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
import numpy as np
import matplotlib.pyplot as plt
x=np.array([
    [4.8,120,200],[3.7,150,100],[4.9,90,300],
    [3.6,170,80],[4.5,110,220]
])
linked=linkage(x,method='ward')
plt.figure(figsize=(10,7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("dendrogram")
plt.xlabel("movie index")
plt.ylabel("euclidean distance")
plt.show()
```
